Understand the principles of prompt engineering and learn to create effective prompts for AI models like ChatGPT. Prompt engineering is crucial for obtaining desired outputs from AI models. It helps in guiding the model's response in a particular direction..

```python
import requests
import json
import os

from config import get_token # Get the access token for the checking APIs below
```

### Checking for foundation-models scenario

```python
## API data
endpoint = "/v2/lm/scenarios" #set the API specific parameters
url =  os.environ["AICORE_BASE_URL"]+endpoint

# Set header
headers = {
  'AI-Resource-Group': variables.RESOURCE_GROUP,
  'Content-Type': 'application/json',
  'Authorization': os.getenv('AICORE_CLIENT_TOKEN')
}
# Run API
api_response = requests.request(
  "GET", 
  url, 
  headers=headers
)
```

```python
# Print result
data = json.loads(api_response.text)
# # Access the configuration list and print each key value details
for i, scenario in enumerate(data["resources"], start=1):
    print(f"\nScenario {i}:")
    for key, value in scenario.items():
        print(f"{key}: {value}")
```

### Creating a configuration

```python
## API data
endpoint = "/v2/lm/configurations" #set the API specific parameters
url =  os.environ["AICORE_BASE_URL"]+endpoint

# Set header
headers = {
  'AI-Resource-Group': variables.RESOURCE_GROUP,
  'Content-Type': 'application/json',
  'Authorization': os.getenv('AICORE_CLIENT_TOKEN')
}

# Set body
payload = {
   "name": "anthropic claude-3.5 sonnet_ssconfig",  # decription of the configuration
    "executableId": "aws-bedrock",
    "scenarioId": "foundation-models",              # create the configuration iwith scenario "foundation-models"
    "versionId":"0.0.1",
    "parameterBindings": [
        {
            "key": "modelName",
            "value": "anthropic--claude-3.5-sonnet" # name of the model
        },
        {
            "key": "modelVersion",
            "value": "latest"
        }
    ],
    "inputArtifactBindings": []
}


# Run API
api_response = requests.post(
  url, 
  headers=headers,
  json=payload
  )
```

```python

print("Status:", api_response.status_code)
print("Headers:", api_response.headers)
print("Body:", api_response.text)
```

### Creating a deployment

Create a deployment with the configuration of anthropic claude-3.5 sonnet created in the previous step

```python
## API data
endpoint = "/v2/lm/deployments" #set the API specific parameters
url =  os.environ["AICORE_BASE_URL"]+endpoint

# Set header
headers = {
  'AI-Resource-Group': variables.RESOURCE_GROUP,
  'Content-Type': 'application/json',
  'Authorization': os.getenv('AICORE_CLIENT_TOKEN')
}

# Set body
payload = {
  "configurationId":  "910d7c54-736b-4dff-b19e-93faf01d68af" # configuration id of "anthropic claude-3.5 sonnet_ssconfig"  
}

# Run API
api_response = requests.post(
  url, 
  headers=headers,
  json=payload
  )
```

```python
# after deployment creation its initial status is waiting to be deployed
print("Status:", api_response.status_code)
print("Headers:", api_response.headers)
print("Body:", api_response.text)
```
